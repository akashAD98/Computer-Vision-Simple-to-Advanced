# Convert an RGB image to Grayscale

In this notebook, I show you how easy you can convert an image space color from RGB to grayscale can be. First of all, make sure you have all the dependencies installed on your system:

- python 3 (I'm using python 3.8)
- Matplotlib
- Numpy

When you have all of this properly installed on your system, you can start working.

Let's import the python modules:

```python
import matplotlib.pyplot
import numpy as np
import sys
```

Now it's time to read our image:

```python
try:
    img = matplotlib.pyplot.imread("image.jpg")
except FileNotFoundError:
    print("=== Image file not found. Please check the path passed to matplotlib.pyplot.imread() ===")
else:
    print("Image loaded!")
    matplotlib.pyplot.imshow(img, cmap=matplotlib.pyplot.get_cmap("gray"))
```

All right, now we can code the conversion algorithm. In this notebook we will use a very simple method of conversion from RGB to gray scale. In this method, the gray intendity value is calculated through a simple averaging of the pixel value in each channel.

$$gray = (R + B + G)/3$$

Thus, we need to go through the entire image to calculate the average value of the values of R, G and B. Using numpy we can easily perform this operation:

```python
#Calculating the average value using numpy and creating a new image in grayscale
gray_img = np.average(img, axis=-1)
```

It's done! Our grayscale image is ready and we can see the result using the imshow command:

```python
matplotlib.pyplot.imshow(gray_img, cmap=matplotlib.pyplot.get_cmap("gray"))
```
